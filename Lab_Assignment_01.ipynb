```python
import nltk
from nltk.tokenize import word_tokenize

# Sample text for demonstration
text = """Natural Language Processing (NLP) is a field of artificial intelligence that focuses on the interaction between computers and human language. It involves tasks such as text classification, sentiment analysis, machine translation, and more. Libraries like NLTK and SpaCy provide powerful tools for these operations."""

# Ensure NLTK data for tokenization is downloaded
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')

print("Original Text:")
print(text)
```

### 1. Count the number of words in the text using NLTK

```python
# Tokenize the text into words
words = word_tokenize(text)

# Count the number of words
word_count = len(words)

print(f"Number of words in the text: {word_count}")
```

### 2. Convert the Entire Text into Lowercase using Python's built-in function

```python
# Convert the entire text to lowercase
lowercase_text = text.lower()

print("Lowercase Text:")
print(lowercase_text)
```
